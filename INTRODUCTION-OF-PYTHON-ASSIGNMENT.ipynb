```python
#64) Write a Python function that checks whether a passed string is palindrome or not
def is_palindrome(string):
    string = string.lower()
    string = ''.join(char for char in string if char.isalnum())
    return string == string[::-1]
```

```python
#75) Write a Python program to read last n lines of a file.
def read_last_n_lines(filename, n):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            if n >= len(lines):
                return "".join(lines)
            else:
                return "".join(lines[-n:])
    except FileNotFoundError:
        return f"Error: File '{filename}' not found."

filename = 'ExampleTextFile.txt'
n = 5
last_n_lines = read_last_n_lines(filename, n)
last_n_lines
```

```python
#82) Write a Python program to copy the contents of a file to another file.
def copy_file_contents(source_file, destination_file):
    try:
        with open(source_file, 'r') as source:
            with open(destination_file, 'w') as destination:
                for line in source:
                    destination.write(line)
        print(f"File '{source_file}' copied to '{destination_file}' successfully.")
    except FileNotFoundError:
        print(f"Error: Source file '{source_file}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


source_file = r'C:\Users\shrut\OneDrive\Desktop\svs.txt'
destination_file = 'destination.txt'
copy_file_contents(source_file, destination_file)
```

```python
#89) How Do You Handle Exceptions with Try/Except/Finally in Python? Explain with coding snippets.
def handle_exceptions():
    try:
        # Code that might raise an exception
        file = open("my_file.txt", "r")  # Attempt to open a file
        # Perform operations on the file
        content = file.read()
        print(content)
    except FileNotFoundError:
        print("Error: File not found.")  # Handle the specific exception
    except Exception as e:
        print(f"An unexpected error occurred: {e}") # Handle other exceptions
    finally:
        if 'file' in locals() and not file.closed: #Check if file exists and is open
          file.close()
          print("File closed.")  # Ensure the file is closed


handle_exceptions()
```

```python
#90) Write python program that user to enter only odd numbers, else will raise an exception.
def get_odd_number():
    while True:
        try:
            num = int(input("Enter an odd number: "))
            if num % 2 != 0:
                return num
            else:
                raise ValueError("Invalid input. Please enter an odd number.")
        except ValueError as e:
            print(e)

# Example usage
odd_num = get_odd_number()
print("You entered:", odd_num)
```
