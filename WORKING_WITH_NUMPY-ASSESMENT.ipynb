```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
import pandas as pd
s = pd.read_csv("historical_automobile_sales.csv")
s
```

```python
#Q 1: Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year.
import matplotlib.pyplot as plt
s.groupby('Year').sum()['Price'].plot(kind='line')
plt.show()
```

```python
#Q 2: Plot different lines for categories of vehicle type and analyze the trend to answer the question Is there a noticeable difference in sales trends
#between different vehicle types during recession periods?
s = pd.read_csv("historical_automobile_sales.csv")
s.groupby('Vehicle_Type').sum()['Price'].plot(kind='line')
plt.show()
```

```python
# prompt: Q 3: Use the functionality of Seaborn Library to create a visualization to comparethe sales trend per vehicle type for a recession period with a non- recession period.
s = pd.read_csv("historical_automobile_sales.csv")
s.groupby('Vehicle_Type').sum()['Price'].plot(kind='bar')
plt.show()
```

```python
#Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period
s.groupby('Vehicle_Type').sum()['Price'].plot(kind='bar')
plt.show()
```
